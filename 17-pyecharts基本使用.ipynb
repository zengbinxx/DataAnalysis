```python
# 导入pyecharts中全部可以绘制的图像
from pyecharts.charts  import *
```

```python
#实例化绘图对象，相当于激活绘图对象
bar = Bar()
# 普通调用
# 添加x轴的数据
bar.add_xaxis(['衬衫','棉衣','毛线衣','袜子','裤子','衣服'])
# 添加y轴数据，第一个参数为系列名称，是必须传入的
bar.add_yaxis('',[10,15,20,21,18,5],color = 'red')
# 显示绘制的图像
bar.render_notebook()
```

```python
# 导入全局配置项
import pyecharts.options as opts
# 链式调用
data1 = [10,15,20,21,18,5]
data2 = [10,15,20,25,30,35]
bar = (
    Bar()
    .add_xaxis(['衬衫','棉衣','毛线衣','袜子','裤子','衣服'])
    .add_yaxis('A',data1,
               #设置数据使用y轴的id
               yaxis_index = 0)
    .add_yaxis('B',data2,yaxis_index = 0)
    #新增坐标轴
    .extend_axis(yaxis=opts.AxisOpts(
    type_='value'))
)
bar.render_notebook()
```

折线图绘制

```python
line = (
    Line()
    .add_xaxis(['2001','2002','2003','2004','2005'])
    .add_yaxis('折线图',[15,14,20,31,16])
)
line.render('折线图.html')
# 在Notebook显示图像
line.render_notebook()
```

```python
import numpy as np
#在1-10之间等差生成20个数
x = np.linspace(1,10,num=20) 
scatter = (
    Scatter()
    .add_xaxis(x)
    .add_yaxis('散点图',2*x**2+1)
)
scatter.render_notebook()
```
