```python
import pandas as pd
# 导入pyecharts的高级设置
import pyecharts.options as opts
from pyecharts.charts import *
# 导入主题配置
from pyecharts.globals import ThemeType
```

```python
# 导入数据
data = pd.read_csv('豆瓣.csv')
data
```

```python
data.info()
```

pd.cut(x, bins, right: bool = True, labels=None)  
x:数据  
bins：整数/列表  如果时整数： 会根据整数将数据划分成相等的若干份
列表：会按照泪飙中的数据来划分  
labels：数据划分后的标签 将数据进行分段


```python
cut_x = data['评论人数']
# 定义表来设置分段
bins = [100000,300000,600000,1000000,1500000,2000000,3000000] # 30w属于10w-30w这个范围,right = True
label = ['10w-30w','30w-60w','60w-100w','100w-150w','150w-200w','200w-300w']
data_cut = pd.cut(cut_x,bins,labels = label)
# 统计数据中有多少哥不同值，并计算每个不同值在列中的个数
data_values = pd.value_counts(data_cut
                             # 关闭排序
                             ,sort = False)
data_values
```

```python
bar = (Bar(init_opts=opts.InitOpts(theme=ThemeType.MACARONS))
      .add_xaxis(label)
      .add_yaxis('柱状图',data_values.values.tolist())
      .set_global_opts(
          title_opts=opts.TitleOpts(title='评论数量区间分布-柱状图'),
          xaxis_opts=opts.AxisOpts(name='评论数'),
          yaxis_opts=opts.AxisOpts(name='电影数量',splitline_opts=opts.SplitLineOpts(is_show=True))
                                  )
      )
bar.render_notebook()
# 如果说绘图之后没有显示图像
# 1.多运行几次，有可能时因为网页还在加载绘图数据，导致一下子没显示出来
# 2.检查代码，注意查看中间变量的数据类型
# 3.去官网查看解决方案
# 4.重启Jupyter， 在绘图
```

```python
wc_data = [[i,j] for i, j in zip(data['电影名称'],data['电影评分'])]
wc_data
```

```python
wc = (WordCloud(init_opts=opts.InitOpts(theme=ThemeType.ESSOS))
     .add('电影名称', wc_data,word_size_range=[5,50])
     .set_global_opts(title_opts=opts.TitleOpts(title='电影名称-评分词云图')
)
     )

wc.render_notebook()
```

```python
# 电影评分和评论人数的折线图x轴：电影名称 y:评分和人数
line = (Line(init_opts=opts.InitOpts(theme=ThemeType.ESSOS))
       .add_xaxis(data['电影名称'][:50].to_list())
       .add_yaxis('电影评分',data['电影评分'][:50].to_list(),yaxis_index=0)
       .add_yaxis('评论人数',data['评论人数'][:50].to_list(),yaxis_index=1) 
        # 新增y轴
       .extend_axis(yaxis=opts.AxisOpts(name='评论人数'))
       .set_global_opts(xaxis_opts=opts.AxisOpts(type_='category',
                                                axislabel_opts=opts.LabelOpts(interval=0,rotate=45)),
                        datazoom_opts = opts.DataZoomOpts(range_start=30,range_end=80),
                        yaxis_opts = opts.AxisOpts(name = '电影评分',
                                                #设置y轴最小值
                                                min_=data['电影评分'].min()-2))
                      
       )
line.render_notebook()
```

```python
# 不同国家电影数量的饼图
data_pie = pd.value_counts(data['国家'])
pie = (
    Pie(init_opts=opts.InitOpts(theme=ThemeType.ESSOS))
    .add('饼图',
         [[i,j] for i,j in zip(data_pie.index.to_list(),data_pie.values.tolist())],
         # 修改饼图的半径
         radius = 150
        )
    .set_global_opts(
        title_opts = opts.TitleOpts(title='国家-电影数量饼图',pos_left='ceter'),
        legend_opts=opts.LegendOpts(type_='scroll',orient='vertical',pos_right=0))
    .set_series_opts(label_opts=opts.LabelOpts(formatter = '{b}:{c}'))
)
pie.render_notebook()
```

```python
bar_y = (
    Bar(init_opts=opts.InitOpts(theme=ThemeType.ESSOS))
    .add_xaxis(data['电影名称'].to_list()[:20])
    .add_yaxis('',data['评论人数'].to_list()[:20])
    .reversal_axis()
    .set_global_opts(opts.TitleOpts(title='电影名称-电影评论数',pos_left='ceter'))
    # 设置图例显示在柱子右侧
    .set_series_opts(label_opts=opts.LabelOpts(position='right')))
bar_y.render_notebook()
```

```python
page = (Page(
    # 配置未拖拽布局
    layout=Page.DraggablePageLayout)
    .add(bar,wc,line,pie,bar_y)
       )
#保存本地
page.render('one.html')
```

```python
# 保存上一次拖拽后的布局
page.save_resize_html(
    # 第一次的html文件（可拖拽的）
    source='one.html',
    # 拖拽后保存生成的json配置文件
    cfg_file='chart_config.json',
    #第二次保存的html文件（不可拖拽）
    dest= '可视化.html'
)
```
