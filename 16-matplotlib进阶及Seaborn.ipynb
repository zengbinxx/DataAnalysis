```python
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 将matplotlib的图表直接嵌入到Notebook中
%matplotlib inline
# 显示中文
plt.rcParams['font.sans-serif'] = 'SimHei' # 只对Windows用户有效，苹果电脑无此字体
# 支持符号
plt.rcParams['axes.unicode_minus'] = False
```

```python
x = [1,2,3]
y = [4,5,6]
plt.bar(x,y)
plt.show()
```

```python
x_datas = ['北京市','上海市','天津市','重庆市']
y_datas = [random.randint(10000,14000) for i in range(4)]
#第一步创建画布，可省略
fig = plt.figure(figsize = (8,4))
#第二步在画布上绘图
plt.bar(x_datas,y_datas,
        #设置柱状图的颜色
        color = 'y',
       #设置柱状图的宽度
       width = 0.6   
       )
#第三步，图例等细节配置，图像显示
plt.title('四个直辖市的GDP')
plt.xlabel('直辖市')
plt.ylabel('GPT')
#修改y轴刻度的范围plt.ylim([最小值，最大值])
plt.ylim([6000,15000])
# 设置柱状图文本标签
# plt.text(x轴坐标，y轴坐标，字符)
for i,j in zip(x_datas,y_datas):
    plt.text(i,j+100,j,
             #设置剧中
            ha = 'center')
# 显示绘制的图像
plt.show()
```

水平条形图  
plt.barh(y,height,width)  
y：表示y坐标，数据类型为int或float类型，刻度自适应调整；也可传dataframe的object，y轴上等间距排列；   
height：表示柱状图的长度，也就是x坐标值，数据类型为int或float类型；  
width：表示柱状图的高度，取值在0~1之间，默认为0.8  
其他参数与plt.bar()类似。

```python
plt.figure(figsize = (12,5))
y_data2 = ['亚马逊','当当网','中国图书网','京东','天猫']
x_data2 = [38.8,40.2,48.4,39.9,32.34]
plt.barh(y_data2,x_data2)
plt.title('不同平台的最低值')
plt.ylabel('平台')
plt.xlabel('价格')
#设置x轴的最小值和最大值
plt.xlim([20,50])
for i,j in zip(x_data2,y_data2):
    plt.text(i+1,j,i,
        #设置居中
        ha = 'center')
plt.show()
```

#### 散点图

plt.scatter(x, y, s=None, c=None, vmin=None, vmax=None, alpha=None, edgecolors=None)

- x,y 形如shape(n,)的数组，可选值
- s 代表点的大小
- c 代表点的填充颜色- marker 点的形状,可选值，默认是圆
- alpha：透明度
- edgecolors 点的边界颜色或颜色序列，可选值，默认值：None  
- vmin:最小值
- vmax：最大值

```python
plt.figure(figsize = (5,3))
height=[120,161,168,175,188,190,200,186,163,158,155,173]
weight=[28,55,50,60,66,56,47,49,60,70,75,63]
plt.scatter(height,weight,s=100,alpha=0.5)
plt.show()
```

饼图

plt.pie(x, explode=None, labels=None, colors=None,autopc)
- x：指定绘图的数据；如果数据小于1，则为百分比，如果大于1则
- explode：指定饼图某些部分的突出显示，即呈现爆炸式；
- labels：为饼图添加标签说明，类似于图例说明；
- colors：指定饼图的填充色；
- autopct：自动添加百分比显示，可以采用格式化的方法显示!

```python
#中专占比25.15%，大专占比37.24%，本科占比33.36%，硕士占比3.68%
data = [0.2515,0.3724,0.3336,0.0368,0.057]
labels = ['中专','大专','本科','硕士','其他']
explod = [0,0.02,0,0,0]
colors = ['red','green','yellow','blue','pink']
plt.pie(data,labels=labels,
       #突出设置
       explode = explod,
       #修改每个饼的颜色
       colors=colors,
       autopct='%1.1f%%',
       #设置百分比标签的圆心距
       pctdistance = 0.3,
       #设置标签的圆心距
       labeldistance = 1.1)
plt.show()
```
