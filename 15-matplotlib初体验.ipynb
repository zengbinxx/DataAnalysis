```python
import numpy as np
x = np.arange(2,20,2)
x
```

折线图的绘制：
语法：
plt.plot(x,y)

```python
y = x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,
         #设置透明度
         alpha=0.5)
plt.show()
```

```python
# 折线样式设置
plt.plot(x,y,
         #设置透明度
         alpha=0.5,
         #设置短点时间线
        linestyle = '--' ,#无法自定义
         # 设置转折点的样式
         marker = 'h',
         markeredgecolor = 'r',#设置外边框颜色为青色
         markerfacecolor = 'r' ,#设置内填充颜色为红色
         markersize = 10 #设置大小
        ) 
plt.title('折线图',fontsize = 20)
plt.xlabel('自变量')
plt.ylabel('因变量')
plt.show()
```

```python
x1 = np.linspace(-np.pi,np.pi,50)#生成三角函数重度参数
plt.plot(x1,np.sin(x1))
plt.xlabel('自变量')
plt.ylabel('因变量')
plt.title('正弦三角函数')
plt.show()
```

```python
x2 = np.arange(2,30,2)
# 随机生成在5—35直接的整数，数据量和x2中的数据量相同
y2 =np.array(np.random.randint(5,35,len(x2)))
# y轴刻度
y_ticks = [f'{i}℃' for i in y2]
# 创建一个大小为20*8的画布
plt.figure(figsize=(20,8))
plt.plot(x2,y2)
# 修改y轴刻度
plt.yticks(y2,y_ticks,fontsize = 10)
plt.title('气温变化图',fontsize = 15)
#保存图片为本地
plt.savefig('气温变化图.jpg')
plt.show()
```

#### 画布相关设置

plt.figure(num = None, figsize = None, facecolor = None, edgecloer = None, frameon = True,FigureClass = <class'matplotlib.figure.Figure'>,clear = False, **kwargs)
num:画布编号，写字符串就是画布名称，如果数字是已经存在的画布编号，那么久表示激活对该画布的引用   
figsize:设置画布的大小，宽度和高度，单位为英寸
dpi:分辨率  
facecolor:画布的背景颜色
edgecolor:边框颜色
frameon:是否显示边框
clear：如果为True并且该图形已经存在，那么就会清空该图形

```python
plt.figure(facecolor='r')
x3 = np.arange(10,15)
plt.plot(x3,x3**2)
plt.show()
```

#### 创建子图

plt.subplot(nrows,ncols,index)  
nrows,ncols:子区域的行数和列数
index：区域的索引
当这几个参数如果都是个位数可以写一起

```python
y3 = np.random.randint(10,15,size = len(x3))
plt.figure(facecolor = 'y')
plt.subplot(221) #将画布分为2*2的区域，并将在编号为1的区域上绘图
plt.plot(x3,x3*2)
plt.subplot(222)
plt.plot(x3,x3**2)
plt.subplot(223)

plt.plot(x3,x3**3)
plt.subplot(224)
plt.scatter(x3,y3)
plt.show()
```

```python
plt.figure(facecolor = 'y')
plt.subplot(221) #将画布分为2*2的区域，并将在编号为1的区域上绘图
plt.plot(x3,x3*2)
# 每一次执行的时候都会对画布进行区域划分
plt.subplot(222)
plt.plot(x3,x3**2)
plt.subplot(212)

plt.plot(x3,x3**3)

plt.show()
```
