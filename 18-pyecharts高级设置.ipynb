```python
x= list(range(10))
scatter = (
    Scatter()
    .add_xaxis(x)
    .add_yaxis('散点图',x)
    .set_global_opts(
        title_opts = opts.TitleOpts(title='这是主标题',subtitle='这是副标题'),
        xaxis_opts=opts.AxisOpts( type_='category',
                                              axistick_opts = opts.AxisTickOpts(is_align_with_label=True),
                                             #取消坐标轴2边的空白区域
                                              boundary_gap=False))
)
scatter.render_notebook()
```

```python
x_data = list(range(2001,2023))
y_data = [random.randint(1,100) for i in x_data]
y_data2 = [random.randint(1,100) for i in x_data]
bar = (
    Bar(init_opts=opts.InitOpts(theme=ThemeType.ESSOS))
    .add_xaxis(x_data)
    .add_yaxis('A',y_data)
    .add_yaxis('B',y_data2)
    .set_global_opts(title_opts=opts.TitleOpts(title='柱状图'),
                     xaxis_opts=opts.AxisOpts(axislabel_opts=opts.LabelOpts(
                         rotate = 45,
                         #强制显示所有标签
                         interval=0))
                    )

)
bar.render_notebook()
```

绘图逻辑:  
1.选择图像类型，无论是matplotlib还是ryecharts这一步是一样  
2.根据官方文档找到自己需要绘制的图像名称  
3.准备数据，不同的图像的数据格式要求不同，使数据满足图像的需求  
4.使用数据进行绘图  
高级设置  
5.观察图像，根据‘全局配置项’中的图来找到自己需要修改的位置  
6.在去‘全局配置项’中找对应的类，并将相应的代码配置好  
7.修改参数

```python
x_data = list(range(2001,2023))
y_data = [random.randint(1,100) for i in x_data]
y_data2 = [random.randint(1,100) for i in x_data]
bar = (
    Bar()
    .add_xaxis(x_data)
    .add_yaxis('A',y_data)
    .add_yaxis('B',y_data2)
    .set_global_opts(datazoom_opts=opts.DataZoomOpts(
        #设置区域缩放的初始位置和结束位置
        range_start=20,end_value=80))

)
bar.render_notebook()
```

```python
# 保存到本地
bar.render('区域缩放.html')
```
